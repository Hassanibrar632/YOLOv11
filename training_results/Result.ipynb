```python
# Load the data into a DataFrame
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./best_fit/results.csv")

# Plot training and validation losses
plt.figure(figsize=(12, 6))
plt.plot(df["epoch"], df["train/box_loss"], label="Train Box Loss", linestyle="dashed")
plt.plot(df["epoch"], df["train/cls_loss"], label="Train Class Loss", linestyle="dashed")
plt.plot(df["epoch"], df["train/dfl_loss"], label="Train DFL Loss", linestyle="dashed")
plt.plot(df["epoch"], df["val/box_loss"], label="Validation Box Loss")
plt.plot(df["epoch"], df["val/cls_loss"], label="Validation Class Loss")
plt.plot(df["epoch"], df["val/dfl_loss"], label="Validation DFL Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and Validation Losses Over Epochs")
plt.legend()
plt.grid(True)
plt.show()

# Plot Precision, Recall, and mAP
plt.figure(figsize=(12, 6))
plt.plot(df["epoch"], df["metrics/precision(B)"], label="Precision")
plt.plot(df["epoch"], df["metrics/recall(B)"], label="Recall")
plt.plot(df["epoch"], df["metrics/mAP50(B)"], label="mAP@50")
plt.plot(df["epoch"], df["metrics/mAP50-95(B)"], label="mAP@50-95")
plt.xlabel("Epochs")
plt.ylabel("Metrics")
plt.title("Precision, Recall, and mAP Over Epochs")
plt.legend()
plt.grid(True)
plt.show()

# Plot Learning Rate Schedule
plt.figure(figsize=(12, 6))
plt.plot(df["epoch"], df["lr/pg0"], label="Learning Rate pg0")
plt.plot(df["epoch"], df["lr/pg1"], label="Learning Rate pg1", linestyle="dashed")
plt.plot(df["epoch"], df["lr/pg2"], label="Learning Rate pg2", linestyle="dotted")
plt.xlabel("Epochs")
plt.ylabel("Learning Rate")
plt.title("Learning Rate Schedule Over Epochs")
plt.legend()
plt.grid(True)
plt.show()

```
